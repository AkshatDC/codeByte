```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error
from pygam import LinearGAM, s

# Load the dataset
file_path = 'input.csv'
df = pd.read_csv(file_path)

# Example: Generalized Additive Model for Fishery Consumption
X1 = df['MPA'].values.reshape(-1, 1)
y1 = df['FisheryConsumption']

# Fit GAM model
gam1 = LinearGAM(s(0)).fit(X1, y1)

# Predictions
XX1 = np.linspace(X1.min(), X1.max(), 100).reshape(-1, 1)
preds1 = gam1.predict(XX1)

# Visualization
plt.scatter(X1, y1, label='Data Points')
plt.plot(XX1, preds1, color='red', label='GAM Fit')
plt.xlabel('MPAs (%)')
plt.ylabel('Fishery Consumption')
plt.legend()
plt.show()

# Example: Generalized Additive Model for Share of global plastics emitted to ocean
X2 = df['MPA'].values.reshape(-1, 1)
y2 = df['Share of global plastics emitted to ocean']

# Fit GAM model
gam2 = LinearGAM(s(0)).fit(X2, y2)

# Predictions
XX2 = np.linspace(X2.min(), X2.max(), 100).reshape(-1, 1)
preds2 = gam2.predict(XX2)

# Visualization
plt.scatter(X2, y2, label='Data Points')
plt.plot(XX2, preds2, color='red', label='GAM Fit')
plt.xlabel('MPAs (%)')
plt.ylabel('Share of global plastics emitted to ocean')
plt.legend()
plt.show()

# Residuals and performance metrics for the second model
residuals2 = y2 - gam2.fittedvalues

plt.scatter(X2, residuals2)
plt.hlines(y=0, xmin=X2.min(), xmax=X2.max(), colors='red')
plt.xlabel('MPAs (%)')
plt.ylabel('Residuals')
plt.show()

r2 = r2_score(y2, gam2.fittedvalues)
mse = mean_squared_error(y2, gam2.fittedvalues)
rmse = np.sqrt(mse)

print(f'R-squared: {r2}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
```
