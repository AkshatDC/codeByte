```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'input.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())

# Calculate correlation matrix
correlation_matrix = data[['MPA', 'FisheryConsumption', 'Share of global plastics emitted to ocean']].corr()
print("Correlation Matrix:\n", correlation_matrix)

# Plotting the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Scatter plot for MPA vs FisheryConsumption
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MPA', y='FisheryConsumption', data=data)
plt.title('MPA vs Fishery Consumption')
plt.xlabel('MPA (%)')
plt.ylabel('Fishery Consumption')
plt.show()

# Scatter plot for MPA vs Share of global plastics emitted to ocean
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MPA', y='Share of global plastics emitted to ocean', data=data)
plt.title('MPA vs Share of Global Plastics Emitted to Ocean')
plt.xlabel('MPA (%)')
plt.ylabel('Share of Global Plastics Emitted to Ocean')
plt.show()
```
